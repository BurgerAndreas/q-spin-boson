```python
import numpy as np

# widget
%matplotlib inline 

from spin_boson import plot_all
```

```python
# initial state
# spin up (excited), one boson
initial = np.kron(np.array([0, 1]), np.array([0, 1, 0]))

# time 0, ..., t
t = 2

# parameters
# epsilon: spin x-field strength
# omega: harmonic oscillator (boson) frequency
# lambd: spin-boson coupling strength
params = [0.1, 1, 1] # epsilon, omega, lambda

plot_all(t, *params, initial)
```

```python
params = [1, 1, 1] # epsilon, omega, lambda
plot_all(t, *params, initial)
```

```python
params = [1, 1, 0.5] # epsilon, omega, lambda
plot_all(t, *params, initial)
```
